Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
titanic_data = pd.read_csv("/tested.csv")
```

```python
titanic_data= titanic_data.drop(columns='Cabin',axis=1)
```

```python
titanic_data['Age'].fillna(titanic_data['Age'].mean(),inplace=True)
```

```python
print(titanic_data['Embarked'].mode())
```

```python
titanic_data['Embarked'].fillna(titanic_data['Embarked'].mode()[0],inplace=True)
```

```python
titanic_data.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
```

```python
titanic_data.head()
```

```python
X = titanic_data.drop(columns = ['PassengerId','Name','Ticket','Survived'],axis=1)
Y = titanic_data['Survived']
```

```python
X.head()
```

```python
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
lireg = LogisticRegression()
```

```python
model = lireg.fit(X_train, Y_train)
```

```python
y_pred = lireg.predict(X_train)
```

```python
print(y_pred)
```

```python
data_accuracy = accuracy_score(Y_train, y_pred)
```

```python
print(data_accuracy)
```

```python
print("Accuracy for training data:",data_accuracy)
```

```python
y_test_pred= model.predict(X_test)
```

```python
print(y_test_pred)
```

```python
test_data_accuracy = accuracy_score(Y_test, y_test_pred)
```

```python
print('Accuracy score of test data : ', test_data_accuracy)
```
